## Drop Id colunm

```python
data = data.drop('ID', axis=1)
```

```python
data.describe()
```

```python
reason_dummies = pd.get_dummies(data['Reason for Absence'], drop_first=True)
```

```python
reason_dummies = reason_dummies.astype(int)
reason_dummies.head()
```

```python
reason_1 = reason_dummies.loc[:,1:14].max(axis=1)
reason_2 = reason_dummies.loc[:,1:14].max(axis=1)
reason_3 = reason_dummies.loc[:,1:14].max(axis=1)
reason_4 = reason_dummies.loc[:,1:14].max(axis=1)
reason_1
```

```python
data = data.drop('Reason for Absence', axis=1)
```

```python
data = pd.concat([data, reason_1, reason_2, reason_3, reason_4], axis=1)
```
